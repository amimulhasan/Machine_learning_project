# remove outlier


```python
q1=df['Administration'].quantile(0.25)
q1
q3=df["Administration"].quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
min_range=q1-(1.5*IQR)
max_range=q1+(1.5*IQR)
min_range,max_range
```

```python
new_df=df[df['Administration']<=max_range]
new_df
```
