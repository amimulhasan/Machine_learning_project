```python
import os
import shutil
import zipfile

# ==============================
# Paths (Kaggle input/output)
# ==============================
source_dir = "/kaggle/input/datasets/amimulhasan/augmented-alzheimer-mri-4-classes-dataset/Augmented Alzheimer MRI 4-Classes Dataset/augmented_data/augmented_data"   # change this
output_dir = "/kaggle/working/cleaned_dataset"
zip_path = "/kaggle/working/cleaned_dataset.zip"

# ==============================
# Classes
# ==============================
classes = [
    "NonDemented",
    "VeryMildDemented",
    "MildDemented",
    "ModerateDemented"
]

# ==============================
# Create output folders
# ==============================
os.makedirs(output_dir, exist_ok=True)

total_count = 0

# ==============================
# Filtering process
# ==============================
for cls in classes:
    src_cls = os.path.join(source_dir, cls)
    dst_cls = os.path.join(output_dir, cls)

    os.makedirs(dst_cls, exist_ok=True)

    count = 0

    for img in os.listdir(src_cls):
        # ❌ skip bracket images
        if "(" in img or ")" in img:
            continue

        shutil.copy2(
            os.path.join(src_cls, img),
            os.path.join(dst_cls, img)
        )
        count += 1
        total_count += 1

    print(f"{cls}: {count} images kept")

print(f"\nTotal cleaned images: {total_count}")

# ==============================
# ZIP the dataset
# ==============================
shutil.make_archive(
    "/kaggle/working/cleaned_dataset",
    'zip',
    output_dir
)

print("ZIP file created at:", zip_path)
```
